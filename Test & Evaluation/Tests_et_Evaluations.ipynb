```python
import re

def evaluate_completeness(query, response):
    completeness_prompt = f"""
    Évalue la complétude de la réponse suivante sur une échelle de 0 à 1 (0 = totalement incomplet, 1 = totalement complet).
    Donne seulement un score numérique suivi de "(complet)" ou "(incomplet)".

    Question : {query}
    Réponse : {response}

    Score :
    """
    llm = ChatGoogleGenerativeAI(google_api_key=GOOGLE_API_KEY, model='gemini-1.5-flash')
    response = llm.invoke(completeness_prompt)

    # Extraire un score numérique dans la réponse
    match = re.search(r"(\d+(\.\d+)?)", response.content)  # Capture un nombre avec un éventuel point décimal
    if match:
        return float(match.group(1))  # Convertit en float
    else:
        raise ValueError(f"Impossible d'extraire un score de complétude dans la réponse : {response.content}")

# Exemple d'utilisation
response_text = "Vous devez vous rendre au Ministère de l'Intérieur et de la Sécurité Publique, faire une demande en fournissant les documents requis, tels qu'une copie de l'acte de naissance sécurisé, un certificat d'identification personnelle, une preuve de profession et une photo d'identité et Payez les frais de dossier de 30 000 francs CFA."


completeness_score = evaluate_completeness("Comment avoir mon passeport?", response_text)
print(f"Completeness Score: {completeness_score:.2f}")
```

```python
def evaluate_faithfulness(context, response):
    faithfulness_prompt = f"Compare le texte suivant avec la source et donne un score entre 0 (pas fidèle) et 1 (très fidèle) :\n\nSource : {context}\n\nRéponse : {response}"

    llm = ChatGoogleGenerativeAI(google_api_key=GOOGLE_API_KEY, model='gemini-1.5-flash')
    response = llm.invoke(faithfulness_prompt)

    # Extraire un score numérique dans la réponse
    match = re.search(r"(\d+(\.\d+)?)", response.content)  # Capture un nombre avec un éventuel point décimal
    if match:
        return float(match.group(1))  # Convertit en float
    else:
        raise ValueError(f"Impossible d'extraire un score de complétude dans la réponse : {response.content}")

# Exemple
retrieved_docs = retriever.get_relevant_documents("Comment avoir mon passeport?")
context_text = "\n\n".join([doc.page_content for doc in retrieved_docs])

faithfulness_score = evaluate_faithfulness(context_text, response_text)
print(f"Faithfulness Score: {faithfulness_score:.2f}")
```

```python
import numpy as np
import pandas as pd
import re
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Exemple de questions pour test
questions = [
    "Comment faire son CIP ?",
    "Comment obtenir un passeport ?",
    "Quels sont les documents nécessaires pour une carte d'identité?",
]

# Réponses de référence (vraies réponses attendues)
reference_answers = [
    "Pour faire votre demande de CIP, munissez-vous des pièces suivantes (en fichier PDF) : Certificat fID / Certificat numéro d’identification personnelle ; Copie de l’ancien acte de naissance ; Attestation de résidence ou facture de SONEB/SBEE accompagné d’un montant de 1000 FCFA.",
    "Vous devez vous rendre au Ministère de l'Intérieur et de la Sécurité Publique, faire une demande en fournissant les documents requis, tels qu'une copie de l'acte de naissance sécurisé, un certificat d'identification personnelle, une preuve de profession et une photo d'identité et Payez les frais de dossier de 30 000 francs CFA.",
    "Vous devez fournir un Acte de naissance sécurisé (obligatoire, peut être obtenu à l'ANIP ou dans ses représentations dans les mairies sur présentation du Récépissé Ravip), une Copie du Certificat d'Identification Personnelle (obligatoire, peut être obtenu à l'ANIP ou dans ses représentations dans les mairies, sur présentation du récépissé RAVIP), Une photo bijou (obligatoire), Preuve de la profession (facultative), Preuve du groupe sanguin (facultative), Quittance ou preuve valable de paiement de 6.000 FCFA sur le compte 102316471 au trésor public (obligatoire)",
]

# Réponses générées (par exemple, par Gemini)
gemini_answers = [
    "Pour obtenir votre Certificat d’Identification Personnelle (CIP) au Bénin, vous devez vous rendre à l'Agence Nationale d’Identification des Personnes (ANIP) avec les documents suivants :Une copie de votre récépissé RAVIP ;Une copie de votre acte de naissance ;Une attestation de résidence ou une facture SBEE/SONEB ;Une quittance de paiement des frais (1800 FCFA).L'ANIP est située à Patte d’Oie, Cadjèhoun Cotonou, à côté de la station SONACOP de Cadjèhoun. Ses heures d'ouverture sont du lundi au vendredi, de 8h00 à 12h30 et de 14h00 à 17h00. Vous pouvez également les contacter par e-mail à contact@anip.bj. Le délai de traitement est estimé à 48 heures.Lien vers le service : Service Public.",
    "Étant donné que vous êtes un Béninois résidant au Bénin et que vous souhaitez obtenir un passeport ordinaire biométrique, vous devez vous rendre sur le site  `https://epasseport.service-public.bj/`.  Le coût est de 30 000 FCFA et la durée de traitement est estimée entre 1 et 2 semaines.  \n\nAvant de commencer la demande en ligne, assurez-vous d'avoir les documents suivants :\n\n* Une copie de votre acte de naissance sécurisé (délivré par l'ANIP)\n* Une copie de votre certificat d'identification personnelle (délivré par l'ANIP)\n* Une preuve de votre profession\n* Une photo d'identité couleur, fond blanc (format PP, normes OACI)",
    "Pour obtenir une carte d'identité béninoise, les documents nécessaires dépendent de la situation du demandeur (première demande, renouvellement, etc.). Cependant, voici les documents généralement requis :\n\n**Pour une première demande :**\n\n* Une copie de l'acte de naissance sécurisé (délivré par l'ANIP)\n* Une copie du certificat d'identification personnelle (délivré par l'ANIP)\n* Preuve de la profession (tout document prouvant la profession)\n* Une (01) photo d'identité couleur fond blanc (format PP normes OACI)",
]

# Fonction pour évaluer la complétude
def evaluate_completeness(query, response):
    completeness_prompt = f"""
    Évalue la complétude de la réponse suivante sur une échelle de 0 à 1 (0 = totalement incomplet, 1 = totalement complet).
    Donne seulement un score numérique suivi de "(complet)" ou "(incomplet)".

    Question : {query}
    Réponse : {response}

    Score :
    """
    llm = ChatGoogleGenerativeAI(google_api_key='AIzaSyCr2BbdHHl5onzxvhbSkPGlumx7NtUGC54', model='gemini-1.5-flash')
    response = llm.invoke(completeness_prompt)

    # Extraire un score numérique dans la réponse
    match = re.search(r"(\d+(\.\d+)?)", response.content)
    if match:
        return float(match.group(1))
    else:
        raise ValueError(f"Impossible d'extraire un score de complétude dans la réponse : {response.content}")

# Fonction pour évaluer la fidélité
def evaluate_faithfulness(context, response):
    faithfulness_prompt = f"Compare le texte suivant avec la source et donne un score entre 0 (pas fidèle) et 1 (très fidèle) :\n\nSource : {context}\n\nRéponse : {response}"

    llm = ChatGoogleGenerativeAI(google_api_key='AIzaSyCr2BbdHHl5onzxvhbSkPGlumx7NtUGC54', model='gemini-1.5-flash')
    response = llm.invoke(faithfulness_prompt)

    # Extraire un score numérique dans la réponse
    match = re.search(r"(\d+(\.\d+)?)", response.content)
    if match:
        return float(match.group(1))
    else:
        raise ValueError(f"Impossible d'extraire un score de fidélité dans la réponse : {response.content}")

# Simulation de l'utilisation de TF-IDF pour obtenir des réponses alternatives
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(reference_answers)

tfidf_answers = []
for question in questions:
    query_vec = vectorizer.transform([question])
    similarities = cosine_similarity(query_vec, X).flatten()
    best_index = np.argmax(similarities)
    tfidf_answers.append(reference_answers[best_index])

# Comparaison des modèles
df_results = pd.DataFrame(columns=["Question", "Complétude (Gemini)",  "Fidélité (Gemini)", "Complétude (TF-IDF)", "Fidélité (TF-IDF)"])

for i in range(len(questions)):
    # Évaluer la complétude et la fidélité pour Gemini
    completeness_g = evaluate_completeness(questions[i], gemini_answers[i])
    faithfulness_g = evaluate_faithfulness(reference_answers[i], gemini_answers[i])

    # Évaluer la complétude et la fidélité pour TF-IDF
    completeness_t = evaluate_completeness(questions[i], tfidf_answers[i])
    faithfulness_t = evaluate_faithfulness(reference_answers[i], tfidf_answers[i])

    # Ajouter une ligne au DataFrame
    new_row = pd.DataFrame({
        "Question": [questions[i]],
        "Complétude (Gemini)": [round(completeness_g, 3)],
        "Fidélité (Gemini)": [round(faithfulness_g, 3)],
        "Complétude (TF-IDF)": [round(completeness_t, 3)],
        "Fidélité (TF-IDF)": [round(faithfulness_t, 3)]
    })
    df_results = pd.concat([df_results, new_row], ignore_index=True)

# Affichage des résultats
print(df_results)
```
